# Exemplo de uso de um algoritmo usando Scipy
- Neste notebook vamos fazer um exemplo de interpolação utilizando a função `interp1d()` da SciPy

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

## Gerando os dados originais
- Vamos gerar dados de uma função cosseno
- A intenção é fazer uma interpolação com esses dados

```python
x = np.linspace(0, 15, num=11)
x
```

```python
y = np.cos(-x**2/9.0)
y
```

## Criando a interpolação
- A função `interp1d()` faz uma interpolação 1D
- Precisamos obrigatoriamente passar uma entrada `x`, uma saída `y`
- Opcionalmente, podemos escolher o tipo da interpolação usando o parâmetro `kind`
    - O padrão é `linear`
    - Mas também vamos usar o `cubic`
- Documentação da função: https://scipy.github.io/devdocs/reference/generated/scipy.interpolate.interp1d.html#scipy.interpolate.interp1d

### Interpolação linear

```python
int_lin = interpolate.interp1d(x, y, kind="linear")
```

### Interpolação cubica

```python
int_cub = interpolate.interp1d(x, y, kind="cubic")
```

### Gerando novos dados

```python
x_test = np.linspace(0, 15, num=50)
```

```python
int_lin(x_test)
```

### Fazendo um plot

```python
plt.plot(x, y, 'o', x_test, int_lin(x_test), '-', x_test, int_cub(x_test), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```
