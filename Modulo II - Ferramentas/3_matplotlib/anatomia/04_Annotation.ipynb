# `Annotation`: criando anotações em um plot
- Uma anotação é quase igual um texto, porém, é possível mesclar texto e um apontador (tipo uma seta)
- Também é utilizado para destacar alguma parte do gráfico
- O seu funcionamento e bem similar. Entretanto, existem diversos parâmetros de configuração

```python
import matplotlib.pyplot as plt
from matplotlib.text import Annotation
```

- **Vamos adicionar uma anotação simples em um plot**
    - Perceba pelo `import` que uma anotação pertence ao subpacote de `text`
    - Sendo assim, elas são adicionadas usando o mesmo método `_add_text()`

```python
annot = Annotation("Importante", 
                   xy=(0.2, 0.2), # coordenadas do ponto que queremos destacar
                   xytext=(0.5, 0.8), # coordenadas do texto
                   arrowprops=dict(arrowstyle='->')) # propriedades da seta (precisa ser passado como dict)

fig, ax = plt.subplots()
ax._add_text(annot)
```

- **Podemos estilizar a seta de várias maneiras**
    - [Documentação](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.patches.FancyArrowPatch.html#matplotlib.patches.FancyArrowPatch)

```python
ax.annotate("Atenção", 
            xy=(0.8, 0.8), 
            xytext=(0.8, 0.1), 
            arrowprops=dict(arrowstyle='<->', lw=2, color='lime') )
fig
```

- **Podemos mesclar com outros atributos e que vimos de texto**
    - Lembresse que anotações e textos são adicionados da mesma forma
    - Logo, acessamos também da mesma forma

```python
ax.texts
```

```python
atencao = ax.texts[1]
atencao
```

- Primeiro vamos aumentar o espaço da figura

```python
fig.set_size_inches((15, 7))
fig
```

- **Vamos estilizar o texto da anotação `atenção`**

```python
# Criando um circulo em volta
atencao.set_bbox(dict(facecolor='gold', boxstyle='circle, pad=1', ls='dashed', lw=2))
atencao.set_size(12)
atencao.set_color('blue')
fig
```

- **Agora vamos customizar a seta da anotação `atenção`**
    - Vamos acessar ela através do atributo `arrow_patch`

```python
seta = atencao.arrow_patch
seta
```

- Perceba que é a mesma coisa que fazer:

```python
ax.texts[1].arrow_patch
```

- Agora vamos estilizar
    - Tuto que é possível fazer está descrito na [documentação](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.patches.FancyArrowPatch.html#matplotlib.patches.FancyArrowPatch)

```python
seta.set_linewidth(2) # alterando a grossura da seta
seta.set_arrowstyle('-[') # alterando o estilo da seta
seta.set_connectionstyle('angle, angleA=-90') # alterando a conectividade
fig
```

```python
seta.set_connectionstyle('arc3, rad=-0.5') # alterando a conectividade
fig
```
