# Labels dos eixos e spines
- Neste notebook vamos dar uma olhada em como customizar os labels dos eixos e os spines
- Spines nada mais é do que as bordas principais do plot

```python
import matplotlib.pyplot as plt
```

- Para isso, vamos criar uma `Figure` e um `Axes` vazio

```python
fig, ax = plt.subplots(figsize=(15,5))
```

- **Podemos obter tudo que constitui o `Axes` da seguinte forma**

```python
ax.get_children()
```

**Em resumo:**
- São quatro spines: top, bottom, left, right
- 2 eixos: X e Y
- 3 textos: título, label do eixo X e Y
- 1 retangulo que é a área em branco do plot


- Perceba que a biblioteca é construída de uma forma muito inteligente! Ela vai empilhando camadas para construir o plot
- Isso da uma flexibilidade de costumização gigante para a lib

## Spines
**Podemos acessar apenas as spines da seguinte forma:**

```python
ax.spines
```

- Perceba que é um dicionário em que cada uma das posições é a chave de uma spine

**Dessa forma, podemos customizar cada uma delas**
- Vamos esconder as bordas do topo e da direita:

```python
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig
```

- Mudar a cor da borda de baixo

```python
ax.spines['bottom'].set_color('r')
fig
```

- Alterar estilo e grossura do da esquerda

```python
ax.spines['left'].set_lw(2)
ax.spines['left'].set_ls('dotted')
fig
```

- Enfim, podemos customizar de muitas formas
- Podemos usar bastante coisa que já foi mostrado nos notebooks anteriores

## Labels
**Podemos acessar os labels dos eixos da seguinte forma:**
- Nota: customizar X ou Y é similar. Então vamos focar mais no X, mas o mesmo vale pra Y

```python
ax.xaxis.get_label()
```

```python
ax.yaxis.get_label()
```

- Podemos trocar o texto do label de duas formas diferentes:

```python
ax.xaxis.set_label_text('Label do eixo X (modo 1)')
fig
```

```python
ax.set_xlabel('Label do eixo X (modo 2)')
fig
```

- Para acessar o label, fazemos:

```python
x_label = ax.xaxis.get_label() # # ou ax.xaxis.label
```

- Agora vamos customizar

```python
x_label.set_size(18) # alterar fonte
x_label.set_fontweight('bold') # deixar negrito
x_label.set_color('red') # alterar cor
x_label.set_x(0.2) # mover a posição
fig
```

- No fundo, o label é `Text`, logo, podemos fazer criar um box como já vimos
    - A intenção aqui é mostrar como que a biblioteca permite a customização

```python
x_label.set_bbox(dict(facecolor='gray', 
                      boxstyle='round, pad=0.2',
                      edgecolor='r',
                      linestyle='dashed',
                      linewidth=1))
fig
```
