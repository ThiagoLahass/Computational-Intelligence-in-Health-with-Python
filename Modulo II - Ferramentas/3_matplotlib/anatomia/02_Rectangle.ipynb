# `Rectangle`: criando retângulos em um plot
- Assim como criamos linhas, também podemos criar retângulos
- Isso é feito através da classe `Rectangle`
- Ideia é similar as linhas, porém, agora temos que trabalhar em cima de um retângulo

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

- **Criando dois retângulos**
    - Novamente, utilize o `shift` + `tab`

```python
ret_1 = Rectangle((0.5, 0), 0.4, 0.8)
ret_2 = Rectangle((0.2, 0), 0.2, 0.6)
```

- **Criando o plot e adicionando os retângulos**
    - No caso das linhas, a gente usavao `add_line`
    - Neste caso, o Retangulo é um `patch` dentro do matplotlib (vide o import)
    - Portanto, vamos adicionar ele usando `add_patch`

```python
fig, ax = plt.subplots()
ax.add_patch(ret_1)
ax.add_patch(ret_2)
```

## Acessando `patches` adicionados em um `Axes`
- De maneira similar as linhas, vamos acessar os `pacthes`

```python
ax.patches
```

- **Como você já deve imaginar, também podemos alterar os atributos**
    - As operações são muito semelhantes as anteriores, então não vamos entrar tanto em detalhes

```python
ax.patches[0].set_facecolor('black')
fig
```

```python
ax.patches[1].set_alpha(0.3)
fig
```

- Fazendo várias alterações em uma mesma barra

```python
ax.patches[0].set_edgecolor('b') # deixando a borda azul
ax.patches[0].set_ls('dashed') # deixando a borda com linha tracejada
ax.patches[0].set_lw(5) # deixando a borda com linha mais grossa
ax.patches[0].set_hatch('\\') # adicionando ranhuras
fig
```

- **Os mesmos atributos de `Figure` e `Axes` também funcionam aqui**

```python
ax.set_title("Um título interessante para meu plot")
ax.set_xlabel('Label do eixo X')
ax.set_ylabel('Label do eixo Y')
fig.set_size_inches(14, 6) 
fig
```
