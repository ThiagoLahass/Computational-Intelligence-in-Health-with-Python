```python
import pandas as pd
```

# Ordenando DataFrames
- Assim como nas séries, podemos usar o método `sort_values()`
- Porém, precisamos informar a coluna para ordenação

```python
nba = pd.read_csv("../data/nba.csv")
nba.head()
```

```python
nba.sort_values(by="Name")
```

```python
nba.sort_values(by="Name", ascending=False)
```

```python
nba.sort_values(by="Salary", ascending=False, inplace=True)
nba
```

- Por padrão, os valores com `NaN` são colocados no final
- Veja o Salary aqui:

```python
nba.tail()
```

- Porem, podemos alterar esse comportamento usando o parametro `na_position`:

```python
nba.sort_values(by="Salary", ascending=False, inplace=True, na_position="first")
nba.head()
```

## Ordenando usando multiplas colunas
- É igualmente fácil, apenas precisamos passar uma lista para o parametro `by` indicando a ordem das colunas a serem consideradas

```python
nba.sort_values(by=["Name", "Salary", "Team"], ascending=[False, True, False], inplace=True)
nba.tail()
```

## Ordenando pelo index
- Assim como para séries, podemos ordenar pelos indices
- Podemos usar o método `sort_index()`

```python
nba.sort_index(ascending=True, inplace=True)
nba
```

# Criando um ranking
- Podemos criar um rank usando o método `rank()`
- Vamos supor que eu queria o ranking de cada jogador de acordo com o salário

```python
nba["Salary"].fillna(0, inplace=True)
nba["Salary"].rank(ascending=False).astype(int)
```

- Agora vamos criar uma nova coluna pra essa informação

```python
nba["Rank Salary"] = nba["Salary"].rank(ascending=False).astype(int)
nba.head()
```

# Obtendo os `n` maiores e menores valores
- Vamos supor que queremos obter os `n` maiores e menores valores de uma coluna
- Podemos resolver isso usando o método `sort_values()`
- Porém, é mais interessante (e muitas vezes mais rápido) usar os métodos `nsmallest()` e `nlargest`
- A ideia é similar ao `sort`, porém, ele retorna apenas os `n` desejados

- Vamos obter os 7 jogadores mais pesados da liga:

```python
nba.nlargest(n=7, columns="Weight")
```

- Assim como no sort, podemos colocar uma segunda coluna para desempate:

```python
nba.nlargest(n=7, columns=["Weight", "Age"])
```

- Agora vamos obter os 7 jogadores mais novos da liga usando o `Weight` como desempate

```python
nba.nsmallest(n=7, columns=["Age", "Weight"])
```

- Uma ultima observação: esses métodos funcionam para um série também:

```python
nba["Age"].nlargest(n=6)
```

# Usando o método `sample` para pegar amostras aleatórias
- O método `sample` retorna uma quantidade de dados de maneira aleatório
    - Podemos usar o parâmetro `n` para um valor exato ou `frac` para uma %

```python
nba.sample(n=2)
```

```python
nba.sample(n=2)
```

- Observe que cada execução resulta em um valor diferente
- também podemos usar frac, que representa a % de 0 a 1.0

```python
nba.sample(frac=0.01) # 1%
```

- Outra opção é setar o `axis` que indica se queremos amostrar linhas ou colunas
    - `axis = 0` -> Linhas
    - `axis = 1` -> Colunas
    - O padrão é 0
- Vamos usar `axis=1` com `n=3` para obter 3 colunas aleatórias:

```python
nba.sample(n=3, axis=1).head()
```

- **Pro tip**: se quisermos embaralhar os dados, podemos usar esse método da seguinte forma:

```python
nba_rand = nba.sample(frac=1.0, random_state=8)
nba_rand.head()
```

- O parametro `random_state` é apenas uma semente para reprodução
