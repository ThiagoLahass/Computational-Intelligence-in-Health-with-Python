# Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt 
```

- Definindo a função

```python
sigmoid = lambda x, a: 1/(1+np.exp(a*-x))
```

- Plotando

```python
x = np.linspace(-10, 10, 500)
y_1 = sigmoid(x, 1)
y_05 = sigmoid(x, 0.5)
y_025 = sigmoid(x, 0.25)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

axs[0].plot(x, y_1)
axs[0].set_title("a = 1")
axs[0].axvline(0, color='k', ls='dotted', alpha=0.25)
axs[0].axhline(0.5, color='k', ls='dotted', alpha=0.25)


axs[1].plot(x, y_05)
axs[1].set_title("a = 0.5")
axs[1].axvline(0, color='k', ls='dotted', alpha=0.25)
axs[1].axhline(0.5, color='k', ls='dotted', alpha=0.25)

axs[2].plot(x, y_025)
axs[2].set_title("a = 0.25")
axs[2].axvline(0, color='k', ls='dotted', alpha=0.25)
axs[2].axhline(0.5, color='k', ls='dotted', alpha=0.25)
```
